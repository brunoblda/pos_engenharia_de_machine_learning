```python
import pandas as pd
```

# Analisando os dados que se encotram no data/raw (dataset_kobe_dev.parquet e dataset_kobe_prod.parque)

```python

# Caminho do arquivo Parquet
df_dataset_kobe_dev = pd.read_parquet("../data/01_raw/dataset_kobe_dev.parquet", engine="pyarrow")  # ou engine="fastparquet"
df_dataset_kobe_prod = pd.read_parquet("../data/01_raw/dataset_kobe_prod.parquet", engine="pyarrow")  # ou engine="fastparquet"

print(df_dataset_kobe_dev.head())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_dev.describe())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_dev.info())
print("________________________________________________________________________________________________")

print()
print("-------------------------------------------------------------------------------------------------")
print()

print(df_dataset_kobe_prod.head())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_prod.describe())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_prod.info())
```

# Tratamento dos dados brutos

```python
# Mantendo somente as colunas [lat, lng, minutes remaining, period, playoffs, shot_distance, shot_made_flag]
df_dataset_kobe_dev = df_dataset_kobe_dev[["lat", "lon", "minutes_remaining", "period", "playoffs", "shot_distance", "shot_made_flag"]]
df_dataset_kobe_prod = df_dataset_kobe_prod[["lat", "lon", "minutes_remaining", "period", "playoffs", "shot_distance", "shot_made_flag"]]

# Verificando se existem valores nulos
rows_with_nulls_kobe_dev = df_dataset_kobe_dev[df_dataset_kobe_dev.isnull().any(axis=1)]
print(rows_with_nulls_kobe_dev)

print("----------------------------------------------------------------------------------------------")

rows_with_nulls_kobe_prod = df_dataset_kobe_prod[df_dataset_kobe_prod.isnull().any(axis=1)]
print(rows_with_nulls_kobe_prod)

print(df_dataset_kobe_dev["shot_made_flag"].value_counts())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_prod["shot_made_flag"].value_counts())
print("________________________________________________________________________________________________")

df_dataset_kobe_dev.dropna(inplace=True)
df_dataset_kobe_prod.dropna(inplace=True)

df_dataset_kobe_dev["shot_made_flag"] = df_dataset_kobe_dev["shot_made_flag"].astype(int)
df_dataset_kobe_prod["shot_made_flag"] = df_dataset_kobe_prod["shot_made_flag"].astype(int)

print(df_dataset_kobe_dev.info())

print()
print("-------------------------------------------------------------------------------------------------")
print()

print(df_dataset_kobe_prod.info())

```

```python
print(df_dataset_kobe_dev.head())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_dev.describe())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_dev.info())
print("________________________________________________________________________________________________")

print()
print("-------------------------------------------------------------------------------------------------")
print()

print(df_dataset_kobe_prod.head())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_prod.describe())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_prod.info())


print(df_dataset_kobe_dev["shot_made_flag"].value_counts())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_prod["shot_made_flag"].value_counts())
print("________________________________________________________________________________________________")

print(df_dataset_kobe_dev["minutes_remaining"].value_counts())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_prod["minutes_remaining"].value_counts())
print("________________________________________________________________________________________________")

print(df_dataset_kobe_dev["shot_distance"].value_counts())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_prod["shot_distance"].value_counts())
print("________________________________________________________________________________________________")

print(df_dataset_kobe_dev["period"].value_counts())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_prod["period"].value_counts())
print("________________________________________________________________________________________________")

print(df_dataset_kobe_dev.isnull().sum())
print("________________________________________________________________________________________________")
print(df_dataset_kobe_prod.isnull().sum())
print("________________________________________________________________________________________________")
```
